Aquí hay una breve explicación de cada columna en el conjunto de datos:

    * **Nombre:** Este campo representa el nombre del paciente asociado al registro de salud.

    * **Doctor:** El nombre del médico responsable de la atención del paciente durante su ingreso.
    * **Hospital:** Identifica el centro de salud u hospital donde fue ingresado el paciente.

    * **Número de Habitación:** El número de habitación donde se alojó el paciente durante su ingreso.
    * **Tipo de Ingreso:** Especifica el tipo de ingreso, que puede ser "Emergencia", "Electivo" o "Urgente", reflejando las circunstancias del ingreso.
    * **Fecha de Alta:** La fecha en que el paciente fue dado de alta del centro de salud, en función de la fecha de ingreso y un número aleatorio de días dentro de un rango realista.
    * **Medicación:** Identifica un medicamento recetado o administrado al paciente durante su ingreso. Ejemplos incluyen "Aspirina", "Ibuprofeno", "Penicilina", "Paracetamol" y "Lipitor".
    * **Resultados de las Pruebas:** Describe los resultados de una prueba médica realizada durante el ingreso del paciente. Los posibles valores incluyen "Normal", "Anormal" o "Inconcluso", indicando el resultado de la prueba.

### Visualizaciones:

- Crear visualizaciones para entender la distribución de las variables (histogramas, gráficos de barras, box plots, etc.).
- Visualizar las correlaciones entre variables (mapa de calor de correlación).


```python
# Seleccionar columnas numéricas
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calcular el número de filas y columnas necesarias
n_cols = 5  # Número de columnas en la cuadrícula
n_rows = (len(numeric_columns) // n_cols) + 1  # Número de filas necesarias

# Crear la figura con el tamaño adecuado
plt.figure(figsize=(20, 5 * n_rows))  # Ajustar el tamaño de la figura

# Crear boxplots para identificar outliers
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(n_rows, n_cols, i)  # Ajustar el número de filas y columnas
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()  # Ajustar el espaciado entre subplots
plt.show()
```

```python
# Histograma de edades
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Análisis de columnas categóricas
categorical_columns = ['Gender', 'Blood Type', 'Medical Condition', 'Admission Type', 'Insurance Provider', 'Medication', 'Test Results']

for col in categorical_columns:
    print(f"Distribución de {col}:")
    print(df[col].value_counts())
    print("\n")
```

```python
# Gráfico de barras para género
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df)
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de barras para tipos de sangre
plt.figure(figsize=(10, 6))
sns.countplot(x='Blood Type', data=df, order=df['Blood Type'].value_counts().index)
plt.title('Distribución de Tipos de Sangre')
plt.xlabel('Tipo de Sangre')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Histograma de montos de facturación
plt.figure(figsize=(10, 6))
sns.histplot(df['Billing Amount'], bins=30, kde=True)
plt.title('Distribución de Montos de Facturación')
plt.xlabel('Monto de Facturación')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Matriz de correlación para variables numéricas
corr_matrix = df[['Age', 'Billing Amount', 'Room Number']].corr()

# Heatmap de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
